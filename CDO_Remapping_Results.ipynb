```python
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
# Open the two NetCDF files
ds1 = xr.open_dataset('file1.nc')
ds2 = xr.open_dataset('file2.nc')

# Extract the 'pre' variable and the first time step
pre1 = ds1['pre'].isel(time=0)
pre2 = ds2['pre'].isel(time=0)

# Plot the first file
ax1 = plt.subplot(1, 2, 1, projection=ccrs.PlateCarree())

pre1.plot(ax=ax1, cbar_kwargs={'shrink': 0.5})


# Add country boundaries to the plot
ax1.add_feature(cfeature.BORDERS, linewidth=2.0, edgecolor='black')

# Add ocean and sea boundaries to the plot
ax1.add_feature(cfeature.OCEAN, color='lightblue')
ax1.add_feature(cfeature.COASTLINE)

# Add a legend to the plot
ax1.legend()

# Plot the second file
ax2 = plt.subplot(1, 2, 2, projection=ccrs.PlateCarree())
pre2.plot(ax=ax2, cbar_kwargs={'shrink': 0.5})

# Add country boundaries to the plot
ax2.add_feature(cfeature.BORDERS, linewidth=2.0, edgecolor='black')

# Add ocean and sea boundaries to the plot
ax2.add_feature(cfeature.OCEAN, color='lightblue')
ax2.add_feature(cfeature.COASTLINE)

# Add a legend to the plot
ax2.legend()

# Show the plot
plt.show()
```
